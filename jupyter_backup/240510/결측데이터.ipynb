```python
import numpy as np
import pandas as pd


# numpy 를 통해 null data 삽입
df = pd.DataFrame([[1, 5, 2], [np.nan, 1, 3], [6, np.nan, 20], [2, 5, np.nan], [2, 5, 2]])
df
```

```python
# true 가 null 값
df.isnull()
```

```python
# 각 컬럼 별로 몇개의 결측치가 있는지 판단
df.isnull().sum()
```

```python
df

#결측치 채우기
df.fillna(1)
```

```python
df.fillna(method="ffill") # 특정 값 채우기 ffill = forward fill 앞에 값으로 대체
```

```python
df.fillna(method="bfill") # 특정 값 채우기 bfill = backward fill 앞에 값으로 대체
```

```python
#결측치 제거
df.dropna()
```
