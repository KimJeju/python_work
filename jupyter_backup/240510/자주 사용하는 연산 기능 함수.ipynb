```python
import pandas as pd

df = pd.DataFrame({
    "가로" : [10,20,30,40,5,6,77,81,95],
    "세로" : [91,84,73,68,5,4,32,20,11],
})

df
```

```python
# 키에 따른 리스트 합산
# axis=1 (행끼리 합산 )
df.sum(axis=1)
```

```python
#열 끼리 합산
df.sum(axis=0)
```

```python
# 리스트의 요소 나누기 인자
df.divide(4)
```

```python
# divide 와 sum을 함께 사용하는 예제
# 열끼리 더한 값을 행으로 나눈다.
df.divide(df.sum(axis=1), axis=0)
```

```python
# 누적곱, 열끼리 누적
df.cumprod(axis=0)
```

```python
# 누적곱의 최대값
df.cummax(axis=1)
```
