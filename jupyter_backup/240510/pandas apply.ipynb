```python
# apply => dataframe 에 파이썬 함수를 적용 할 수 있다.

import pandas as pd

df = pd.DataFrame({
    "가로" : [10,20,30,40,5,6,77,81,95],
    "세로" : [91,84,73,68,5,4,32,20,11],
})

df
```

```python
def getVolume(row):
    return row["가로"] * row["세로"]


#하나의 행이 인자가 되어 계산을 수행한다.
df['부피'] = df.apply(getVolume, axis=1)
df
```

```python
df2 = pd.DataFrame({
    "가로" : [10,20,30,40,5,6,77,81,95],
    "세로" : [91,84,73,68,5,4,32,20,11],
})

pd.concat([df,df2], axis=1)
```
