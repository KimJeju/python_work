```python
import pandas as pd
import numpy as np
```

```python
sample = pd.read_csv('data/내국인 생활인구.csv', encoding='cp949')

# 컬럼명만 가져오기
sample.columns
```

```python
#한글명만 가져오기
newCols = []
cols = sample.columns

for col in cols:
    newCols.append(col.split("(")[0])

newCols
```

```python
#뽑아온 한글컬럼을 다시 샘플 테이블에 컬럼명으로 지정해준다.
sample.columns = newCols
sample.head()
```

```python
# 일자를 인덱스로 지정
sample = sample.set_index("일자")
sample.head()
```
